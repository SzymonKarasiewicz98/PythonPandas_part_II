Druga cześć projektu wykorzystująca możliwości bibiolteki Pandas

```python
import pandas as pd

p='./Ceny_produktow_cleaned.csv'
df2=pd.read_csv(p, sep=';', decimal='.')
df=df2.copy()
df.head(10)
```

```python
df['value'].describe() #sprawdzamy wartości w kolumnie value i widzmy, że max to 3000 a min to 0. Warto się temu przyjrzeć
```

```python
df.query('value == 3000') #widzimy, że musi być to błąd, dlatego warto wykluczyć te dane z dalszej analizy
```

```python
w1 = (df['value'] != 3000) & (df['value'] != 0) #wyrzucamy wartości 0 oraz 3000
df=df[w1]
```

```python
df['value'].describe() #sprawdzamy czy filtr zadziałał
```

Pokaże teraz dwa sposoby na pokazania danych grupowanych. Pierwszy sposób przy użyciu groupby oraz drugi przy stworzeniu Pivota

```python
df_grouped=df.groupby(['product', 'year'])['value'].mean()
df_grouped
```

```python
pivot = pd.pivot_table(data=df, index=['product'], columns=['year'], values=['value'], aggfunc='mean')
pivot.head(1) #pokazuję tylko 1 product ze względu na wielkość tego pivota. Jak widać wartości w metodzie group by i pivot table są takie same. 
```

Dodatkowo widzimy, że mamy plik dims, w którym znajdują się 3 arkusze. W 1 mamy province id oraz nazwe province. W 2 mamy product id, product oraz product group id. W 3 mamy product group id oraz product group 

```python
product = pd.read_excel(r'dims.xlsx', sheet_name=1) #zamiast znać nazwę arkusza i ją podawać, używam liczb. Indexowania zaczyna się w Pythonie od 0 dlatego dla arkusza 2 index to 1.
product_group=pd.read_excel(r'dims.xlsx', sheet_name=2) #a dla arkusza nr3 index do 2
```

```python
df.head(5) #szybki podgląd danych aby wiedzieć jak połączyć tabele
```

```python
df_merged=df.merge(product, on='product', how='left') #najpierw mergujemy nasz DF po product, aby otrzymać product_group_id
df_merged
```

```python
df_merged=df_merged.merge(product_group, on='product_group_id', how='left') #jeszce raz mergujemy i uzyskuje w ten sposób kolumnę product_group
df_merged
```

```python
pivot = pd.pivot_table(df_merged, index=['year'], columns=['product_group'], values=['value'], aggfunc=['mean', 'median']) 
pivot
#dzięki wcześniejszemy mergowaniu jesteśmy teraz w stanie stworzyć pivota pokazującego średnią oraz medianę dla każdej grupy produktów na przestrzeni lat 
```
